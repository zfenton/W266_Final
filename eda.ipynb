```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv("Table_1_Exploratory_Data_With_Scripts.csv")
df.head()
```

```python
len(df)
```

```python
# remove rows where the script is missing
df = df.dropna(subset=['script'])
# Remove rows where the script is less than 100 characters
df = df[df['script'].str.len() > 100]
#remove duplicates of the script
df = df.drop_duplicates(subset=['script'])
```

```python
df.reset_index(drop=True, inplace=True)
len(df)
```

```python
df.head()
```

```python
avg_script_length = df['script'].apply(lambda x: len(x.split())).mean()
min_script_length = df['script'].apply(lambda x: len(x.split())).min()
max_script_length = df['script'].apply(lambda x: len(x.split())).max()
print(f"Average script length: {avg_script_length}\n Min script length: {min_script_length}\n Max script length: {max_script_length}")
```

```python
# create new dataframe called df_test with only the script and bechdel_rating columns
df_test = df[['script', 'bechdel_rating']]
```

```python
# Change bechdel_rating so 0, 1, and 2 are all 0 and 3 is 1
df_test['bechdel_rating'] = df_test['bechdel_rating'].replace([0, 1, 2], 0)
df_test['bechdel_rating'] = df_test['bechdel_rating'].replace(3, 1)
```

```python
df_test.head()
```

```python
df_test.to_csv('df_test.csv', index=False)
```
